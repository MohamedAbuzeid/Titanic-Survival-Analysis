```python
import pandas as pd
import numpy as np
from numpy.random import randn
np.random.seed(42)
```

```python
df=pd.read_excel('titanic3.xls')
df
```

```python
df.drop(['home.dest','name','fare','embarked','ticket','parch','sibsp'],axis=1,inplace=True)
df.head(21)
```

```python
age_cat=[]
for age in df['age']:
    if 0 <= age <= 18:
        age_cat.append('child')
    elif 19 <= age <= 30:
        age_cat.append('young adult')
    elif 31 <= age <= 40:
        age_cat.append('adult')
    elif 41 <= age <= 50:
        age_cat.append('senior')
    else:
        age_cat.append('old')

df['age_category']=age_cat
df['age_category']
```

```python
# Iterate through each unique combination of gender, class, and age category
for gender in df['sex'].unique():
    for pclass in df['pclass'].unique():
        # Calculate the total number of passengers in this gender and class
        total_passengers = df[(df['pclass'] == pclass) & (df['sex'] == gender)]['pclass'].count()
        
        # Calculate the total number of survivors in this gender and class
        total_survivors = df[(df['pclass'] == pclass) & (df['sex'] == gender) & (df['survived'] == 1)]['pclass'].count()
        
        # Calculate the survival percentage for this gender and class
        survival_percentage = (total_survivors / total_passengers) * 100 if total_passengers > 0 else 0
        
        print(f'Total survivors from {gender} in class {pclass} is {total_survivors}')
        print(f'Survival percentage from {gender} in class {pclass} is {survival_percentage:.1f} %')
        
        # Now iterate through each age category for this gender and class
        for age_cat in df['age_category'].unique():
            # Calculate the number of survivors in this age category
            survivors_age_cat = df[(df['pclass'] == pclass) & (df['sex'] == gender) & (df['age_category'] == age_cat) & (df['survived'] == 1)]['age_category'].count()
            
            print(f'Survivors from {gender} in class {pclass} and age category {age_cat} is {survivors_age_cat}')
        
        print('\n======================\n')
```
